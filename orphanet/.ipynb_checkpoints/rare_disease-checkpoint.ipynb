```python
##List of disease and Orphacode from Orphanet database from xml file version 01-07-24 (https://www.orphadata.com/genes/)
##Try with ordo id from disgenet
##mail disgnet - get list of ordo annotated disease

##try with list of genes from orpahnet itself - first try this

from bs4 import BeautifulSoup
import pandas as pd


with open('orphanet-list.xml', 'r') as f:
    data_xml = f.read()

soup = BeautifulSoup(data_xml, "xml")

# Initialize an empty list to store the data
data = []  # Ensure this is a list

disorders = soup.find_all('Disorder')

# Extract OrphaCode and Name with lang="en" for each Disorder
for disorder in disorders:
    orpha_code = disorder.find("OrphaCode").text  # Extract OrphaCode
    name_tag = disorder.find("Name", {"lang": "en"})  # Find Name with lang="en"
    omim_id = disorder.find("")
    if name_tag:
        disease_name = name_tag.text
        # Append the data as a tuple
        data.append((orpha_code, disease_name))


 # Create a DataFrame
df = pd.DataFrame(data, columns=["OrphaCode", "Disease_Name"])

# Display the DataFrame
print(df)       
```

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(data_xml, 'xml')

    # Find all disorders
disorders = soup.find_all('Disorder')
disease_info = []

# Loop through each disorder
for disorder in disorders:
        disorder_name = disorder.find('Name', lang="en").text
        disorder_id = disorder['id']
        orpha_code = disorder.find('OrphaCode').text
        
        gene_association = disorder.find('Gene')
        if gene_association:
            gene_name = gene_association.find('Name', lang="en").text
            gene_symbol = gene_association.find('Symbol').text
            
            # Debug External References
            external_references = gene_association.find_all('ExternalReference')
            ensembl_ref = None

            # Search for Ensembl source
            for ext_ref in external_references:
                source = ext_ref.get('Source', '')
                if source == "Ensembl":
                    ensembl_ref = ext_ref.find('Reference').text if ext_ref.find('Reference') else "N/A"
                    break
            
            # Append the data
            disease_info.append({
                "Disorder Name": disorder_name,
                "OrphaCode": orpha_code,
                "Disorder ID": disorder_id,
                "Gene Name": gene_name,
                "Gene Symbol": gene_symbol,
                "Ensembl Reference": ensembl_ref or "N/A"  # Handle missing Ensembl Reference
            })
    
df = pd.DataFrame(disease_info)

# Display the DataFrame
print(df)      
```

```python
import pandas as pd
import numpy as np


import py4cytoscape as p4c
p4c.cytoscape_ping()

import os
print('Get current working directory : ', os.getcwd())

from IPython.display import Image
from IPython.display import SVG

pd.set_option('display.max_colwidth', 1)


# Extract unique IDs for genes and KEs
unique_orpha_id = pd.unique(df[['OrphaCode']].values.ravel())
unique_gene_id = pd.unique(df[['Gene Symbol']].values.ravel())

# Combine the unique IDs
combined_ids = np.concatenate([unique_orpha_id, unique_gene_id])

nodes = pd.DataFrame({'id': combined_ids})
    # Create edges dataframe
edges = pd.DataFrame(data={'source': df['OrphaCode'], 'target': df['Gene Symbol'], 'interaction': 'interacts', 'weight': 1 })
print(nodes)  
    # Create network
p4c.create_network_from_data_frames(nodes, edges, title= "orphanet network")
```
